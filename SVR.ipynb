Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[: , 1:-1].values
y = dataset.iloc[: , -1].values
#y = y.reshape(len(y),1)
y = y.reshape(-1,1)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

Training the SVR model

```python
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X, y)
```

Predicting on a specific value

```python
sc_y.inverse_transform(regressor.predict(sc_X.transform([[6.5]])).reshape(-1,1))
```

Visualising the regrosser

```python
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), c='red')
plt.plot(sc_X.inverse_transform(X), sc_y.inverse_transform(regressor.predict(X).reshape(-1,1)))
plt.title('Prediction of Salary - SVR')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

Visualising in higher resolution

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(-1,1)
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), c='red')
plt.plot(sc_X.inverse_transform(X_grid), sc_y.inverse_transform(regressor.predict(X_grid).reshape(-1,1)))
plt.title('Prediction of Salary - SVR')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#Hadelin de Ponteves:
X_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'red')
plt.plot(X_grid, sc_y.inverse_transform(regressor.predict(sc_X.transform(X_grid)).reshape(-1,1)), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
