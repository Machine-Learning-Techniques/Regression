Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[: , 1:-1].values
y = dataset.iloc[: , -1].values
```

Training the Polynomial Linear Regression model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_f = PolynomialFeatures(degree=4)
X_poly = poly_f.fit_transform(X)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_poly, y)
```

Visualising the regressor

```python
plt.scatter(X, y, c='red')
plt.plot(X, regressor.predict(X_poly), c="blue")
plt.title('Prediction of Salary')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

Visualising in a smoother way

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, c='red')
plt.plot(X_grid, regressor.predict(poly_f.fit_transform(X_grid)), c="blue")
plt.title('Prediction of Salary')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

Predicting on a specific value

```python
int(regressor.predict(poly_f.fit_transform([[6.5]])))
```
